```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float32)
```

```python
pipe = pipe.to("cuda")
```

```python
from PIL import Image

def image_grid(imgs, rows, cols):
    assert len(imgs) == rows * cols

    w, h = imgs[0].size
    grid = Image.new('RGB', size=(cols * w, rows * h))
    grid_w, grid_h = grid.size

    for i, img in enumerate(imgs):
        grid.paste(img, box=(i % cols * w, i // cols * h))
    return grid
```

```python
prompts = ["a photograph of a chef",
           "a photograph of a cook", "a photograph of a firefighter" ,
           "a photograph of a flight attendant",
           "a photograph of a housekeeper",
           "a photograph of a nurse",
           "a photograph of a pilot",
           "a photograph of a software developer",
           "a photograph of a taxi driver",
           "a photograph of a therapist"]
```

```python
# Define the prompt and number of images to generate
num_images = 10

# Generate and save 10 images with the same prompt
for i in prompts:
  prompt = i * num_images
  prompt_identifier = i

  for idx in range(1, 11):
      images = pipe(prompt).images

    # Create an image grid and save it
      grid = image_grid(images, rows=1, cols=1)
      grid.save(f'/content/drive/My Drive/Images/{prompt_identifier}_{idx}.png')
```
